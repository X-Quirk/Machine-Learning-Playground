```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from jupyterthemes import jtplot
jtplot.style()
```

```python
# Reading csv data

homePrice = pd.read_csv('C:/Users/LENOVO/# Jupyter Notebook Files/Data Sets/House Pricing/homeprice.csv')
homePrice
```

```python
# Viewing the data

plt.scatter(homePrice.area,homePrice.price,marker='x',color='cyan',s=80)
```

```python
# Training the model

model = linear_model.LinearRegression()
model.fit(homePrice[['area']],homePrice.price)
```

```python
# Plotting the mathematical model 

plt.scatter(homePrice.area,homePrice.price,marker='x',color='cyan',s=80)
plt.plot(homePrice.area,model.predict(homePrice[['area']]),color='lightgreen')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Predicting the output

area = float(input("Enter the size of the plot you are looking for : "))
price = int(model.predict([[area]]))
print("You would problaby get a plot of {} sq ft at a aprrox. price of $ {}".format(area,price))
```

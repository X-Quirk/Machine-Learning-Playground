```python
import numpy as np
import pandas as pd
from jupyterthemes import jtplot
jtplot.style()
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
# Loading the Digits Data set
# load_digits data set contains 1797, 8x8 pixel images of handwritten digits 
digits = load_digits()
```

```python
dir(digits)
```

```python
# Viewing an image, first image of the data set 
digits.data[0]
# It displays a 1D array of 8x8 pixels so, 64 elements in total
```

```python
# Displaying a sample image from the data set
plt.gray()
plt.matshow(digits.images[0])
# Look closely to identify the digit '0' 
```

```python
# Displaying more sample images from the data set
plt.gray()
for i in range(10):
    plt.matshow(digits.images[i])
# Look closely to identify the fisrt 10 digits i.e 0 to 9 
```

```python
# Viewing the output of respective image
digits.target[0:5]
# It basically means, image 1 displays the digit 0, image 2 displays the digit 1 and so on ...
```

```python
# Splitting training and testingdata sets
X_Train,X_Test,y_Train,y_Test = train_test_split(digits.data,digits.target,test_size=0.4)
```

```python
# Creating the model
model = LogisticRegression()
```

```python
# Training the model
model.fit(X_Train,y_Train)
```

```python
# Predicting the accuracy of the model
model.score(X_Test,y_Test)
```

```python
# Predicting the Output
plt.matshow(digits.images[97])
# Can't really says what digit it is, too pixalted it's 8x8 
# Looks like a 4 though
```

```python
# Checking what digit it actually is
digits.target[97]
# It was actually 4
```

```python
# Predict the output now
model.predict(digits.data[[97]])
# Output satisfied, actually predicted 4
```

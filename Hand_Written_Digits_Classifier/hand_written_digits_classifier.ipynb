# Hand Written Digits Classifier

This is a simple Classifier which classifies Hand Written Digits into the actual digits using a Support Vector Machine (SVM) model based the given input data set which would be a set of images of hand written digits in 8x8 pixels form.

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style()
```

# Loading Data

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
plt.gray()
for i in range(0,10):
    plt.matshow(digits.images[i]) 
    plt.xlabel(digits.target_names[i])
    plt.show() 
    
```

# Data Cleaning

* No Discrepencies were found in the data set

# Train Test Data set Split

```python
from sklearn.model_selection import  train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(digits.data,digits.target,test_size=0.4)
```

   # Model Creation

```python
from sklearn.svm import SVC
```

```python
digits_classifier = SVC()
```

# Training the Model

```python
digits_classifier.fit(x_train,y_train)
```

# Testing the Model

```python
digits_classified = digits_classifier.predict(x_test)
digits_classified
```

# Accuracy of the Model

```python
score = digits_classifier.score(x_test,y_test)
print("The Model classified the Hand Written Digits into actual digits with an accuracy of {} % ".format(int(score*100)))
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix_values = confusion_matrix(y_test,digits_classified)
confusion_matrix_values
```

```python
import seaborn as sns

sns.heatmap(confusion_matrix_values,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

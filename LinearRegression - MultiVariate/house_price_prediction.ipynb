```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from jupyterthemes import jtplot
jtplot.style()
```

```python
# Reading csv data

housePrice = pd.read_csv('C:/Users/LENOVO/# Jupyter Notebook Files/Data Sets/House Pricing/data.csv')
```

```python
# Structing data

housePrice = housePrice[['sqft_living','floors','waterfront','yr_built','bathrooms','view','price']].copy()
housePrice = housePrice.rename(columns={'sqft_living':'area'})
housePrice
```

```python
# Training the model

model = linear_model.LinearRegression()
model.fit(housePrice[['area','floors','waterfront','yr_built','bathrooms','view']],housePrice.price)
```

```python
# Plotting the mathematical model
# Not much useful in plotting the model as the model would be in atleast 6 Dimensions
```

```python
# Predicting the output

area = float(input("What would be the approx. area in sq ft of the house you are looking for ? : "))
floors = float(input("How many floors would you like ? :"))
waterfront = input("Would you prefer a house with a waterfront ? ").lower()
pos = ['yes','y','yeah','why not','yea','y not','yup','definitely','positive','1']
if waterfront in pos:
    waterfront = 1
else:
    waterfront = 0
bathrooms = float(input("How many bathrooms would you prefer for your house ? : "))
yrBuilt = int(input("Which year of make/built would you prefer ? : "))
view = int(input("On a rating of 0 to 5, how would youu want to rate the view of your house ? : "))
price = int(model.predict([[area,floors,waterfront,bathrooms,yrBuilt,view]]))
print("\nYou would most likely get a house with those specifications at an approx. price of $ {}.".format(price))
```
